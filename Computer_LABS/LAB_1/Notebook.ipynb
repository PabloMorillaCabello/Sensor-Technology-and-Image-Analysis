# **Simple Example: Basic Imaeg Operations**

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("test.jpg")##test.jpg should be in the same folder asyour *. ipynb file

# Check if image loaded
if img is None:
    raise ValueError("Image not found. Place test.jpg in the same folderas your notebook.")

# Display image (convert BGR RGB)
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()

# Print image information
print("Image shape:", img.shape) # (H, W, C)
print("Data type:", img.dtype)
print("Min/Max values:", img.min(), img.max())

# Take a region of interest (ROI)
roi = img[300:400 , 350:450]
plt.imshow(cv2.cvtColor(roi , cv2.COLOR_BGR2RGB))
plt.title("ROI")
plt.axis("off")

plt.show()

# Modify pixel values: draw a red square
img [300:400 , 350:450] = [0, 0, 255] # BGR format

# Display modified image
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
plt.title("Modified Image")
plt.axis("off")
plt.show()

# Save result
cv2.imwrite("modified_output.jpg", img)
print("Saved: modified_output.jpg")
```
